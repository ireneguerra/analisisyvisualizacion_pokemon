```python
import pandas as pd

combat = pd.read_csv('combats.csv')
pokemon = pd.read_csv('pokemon.csv')
```

```python
import numpy as np

pokemon = pokemon.rename(index=str, columns={"#": "Number"})
numberOfWins = combat.groupby('Winner').count()
countByFirst = combat.groupby('Second_pokemon').count()
countBySecond = combat.groupby('First_pokemon').count()
numberOfWins = numberOfWins.sort_index()
numberOfWins['Total Fights'] = countByFirst.Winner + countBySecond.Winner
numberOfWins['Win Percentage']= numberOfWins.First_pokemon/numberOfWins['Total Fights']
results = pd.merge(pokemon, numberOfWins, left_on='Number', right_index = True, how='left')
results[results['Win Percentage'].isnull()]
results = results.drop(['First_pokemon', 'Second_pokemon', 'Total Fights'], axis=1)
results[np.isfinite(results['Win Percentage'])]
results['Total'] = results['HP'] + results['Attack'] + results['Defense'] + results['Sp. Atk'] + results['Sp. Def'] + results['Speed']
results.insert(10, 'Total', results.pop('Total'))
results.head()
```

```python
results.to_csv('pokemon_data.csv')
```
